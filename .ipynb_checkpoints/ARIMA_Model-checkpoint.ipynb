```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime , timedelta
import seaborn as sns
sns.set()
```

## Downloading the historical data of a Stock

```python
ticker_symbol = "MSFT"
```

```python
start_date = datetime(year=2019, month=1, day=1)  # Specify January 1st, 2019
end_date = datetime(year=2023, month=1, day=1)  # Specify January 1st, 2023
```

```python
# Download historical data
msft_data = yf.download(ticker_symbol, start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))
```

```python
msft_data.head()
```

```python
df = msft_data.copy()
```

## Data Cleaning and Exploration

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Visualize closing price time series
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.plot(df['Close'], label='MSFT Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Microsoft Stock Closing Price (Jan 2019 - Jan 2023)')
plt.legend()
plt.grid(True)
plt.show()
```

##  Check for Stationarity

```python
# KPSS Test
```

```python
from statsmodels.tsa.stattools import kpss , adfuller
```

```python
stats, p, lags, critical_values = kpss(df['Close'])
```

```python
print(f'Test Statistics: {stats}')
print(f'p-value: {p}')
print(f'Critial Values: {critical_values}')

if p < 0.05 :
  print('Series is not Stationary')
else:
  print('Series is Stationary')
```

```python
# ADF Test 
```

```python
result = adfuller(df['Close'])
print(f'Test Statistics: {result[0]}')
print(f'p-value: {result[1]}')
print(f'Critial Values: {result[4]}')

if result[1] > 0.05 :
  print('Series is not Stationary')
else:
  print('Series is Stationary')
```

## Data Preprocessing

```python
# Create a differenced series (d=1)
differenced_data = df['Close'].diff(1).dropna()
differenced_data
```

```python
# Checking for Stationarity

stats, p, lags, critical_values = kpss(differenced_data)
print(f'Test Statistics: {stats}')
print(f'p-value: {p}')
print(f'Critial Values: {critical_values}')

if p < 0.05 :
  print('Series is not Stationary')
else:
  print('Series is Stationary')
```

```python
result = adfuller(differenced_data)
print(f'Test Statistics: {result[0]}')
print(f'p-value: {result[1]}')
print(f'Critial Values: {result[4]}')

if result[1] > 0.05 :
  print('Series is not Stationary')
else:
  print('Series is Stationary')
```

```python
# Visualize closing price time series
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.plot(differenced_data, label='MSFT Closing Price with differencing')
plt.xlabel('Date')
plt.ylabel('Closing Price with differencing')
plt.title('Microsoft Stock Closing Price (Jan 2019 - Jan 2023)')
plt.legend()
plt.grid(True)
plt.show()
```

### Plotting the ACF and PACF plots

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf , plot_pacf
```

```python
plot_acf(differenced_data , lags=25)
plt.show()
```

```python
plot_pacf(differenced_data , lags=25)
plt.show()
```

## Splitting the data into train and test data

```python
# Split data into training and testing sets (common practice is 80% training, 20% testing)
train_size = int(len(differenced_data) * 0.8)  # Adjust split ratio if needed
train_size
```

```python
train, test = differenced_data.iloc[:train_size], differenced_data.iloc[train_size:]
```

```python
train = train.asfreq('D')
train = train.interpolate('linear')
train
```

```python
test = test.asfreq('D')
test = test.interpolate('linear')
test
```

## Fitting the model

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(train, order=(1, 1, 1)) 
model_fit = model.fit()
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
# Forecast on test data

predictions = model_fit.predict(start=len(train), end=(len(train) + len(test)))
```

```python
predictions
```

```python
test
```

```python
# 2. Calculate Root Mean Squared Error (RMSE)
rmse = mean_squared_error(test, predictions[:-1], squared=False)  # Note: squared=False for direct RMSE calculation
print(f"RMSE: {rmse}")

# 3. Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(test, predictions[:-1])
print(f"MAE: {mae}")
```

```python
residuals = model_fit.resid
plot_acf(residuals)
plt.show()
```

```python
plot_pacf(residuals)
plt.show()
```

## Forecasting the stock Price using the trained model

```python
forecast_horizon = len(test)
forecast = model_fit.forecast(steps=forecast_horizon)
forecast
```
